# Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def extract(n):
    title=[]
    comp=[]
    location=[]
    Exp=[]
    link=[]
    pos=[]
    
    for i in range(2,n+2):
        
        # Creating beatifulsoup object
        #r = requests.get("https://www.shine.com/job-search/data-analyst-jobs-3?q=data%20Analyst".format(i))
        r = requests.get("https://www.shine.com/job-search/data-analyst-data-engineerdata-scientist-python-developer-jobs-{}?q=Data%20Analyst,%20Data%20Engineer,Data%20Scientist,%20Python%20Developer,".format(i))
        
        soup=BeautifulSoup(r.text,'html.parser')

        ##Extracting title of Job description & Apply link
        t=soup.find_all('h2',attrs={"itemprop":"name"})
        for i in t:
            title.append(i.text)
            
        for i in t:
            a=i.find('a')
            l='https://www.shine.com'+a.get('href')[:-1]
            link.append(l)
            
        ##Extracting company name of Job description   
        c=soup.find_all('div',attrs={'class':"jobCard_jobCard_cName__mYnow"})
        for i in c:
            comp.append(i.text)
            
        ##Extracting loctaion & experience required of Job description   
        
        l=soup.find_all('div',attrs={'class':"jobCard_jobCard_lists__fdnsc"})
        for i in l:
            x=i.text.split('Yrs')
            if len(x)>1:
                location.append(x[1])
                Exp.append(x[0])
                
            elif len(x)==1 and re.search('\d',x[0]):
                location.append("Not Defined")
                Exp.append(x[0])
            else:
                Exp.append("Not Defined")
                location.append(x[0])

                 
        ##Extracting number of openings       
        o=soup.find_all('ul',attrs={'class':"jobCard_jobCard_jobDetail__jD82J"})
        for i in o:
            op=i.text
            x=''
            for j in op:
                if j.isdigit():
                    x+=j
            if len(x)>0:
                pos.append(int(x))
            else:
                pos.append(1)
        
    return title,comp,Exp,pos,location,link
```

```python
no=int(input("Enter number of pages to be scrap: "))
```

```python
val=extract(no)
```

```python
df=pd.DataFrame({'Job Title':val[0],'Company Name':val[1],'Exp':val[2],'Number of opening':val[3],'Location':val[4],'Link':val[5]})
```
