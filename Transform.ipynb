## Converting Experience column in required form

```python
df["Exp"]=df['Exp'].apply(lambda x:x.split()[0])
```

## Category wise classification of job description 

```python
def check_title(n):
    n=' '.join(list(map(lambda x:x.lower(),n.split())))
    if re.search('business ana',n):
        return "Business Analyst"
    elif re.search('analy',n):
        return "Data Analyst"
    elif re.search('scientist',n):
        return "Data Scientist"
    elif re.search('engineer',n):
        return "Data Engineer"
    elif re.search('python',n):
        return "Python Developer"
    else:
        return "Other"
```

```python
df["Job Category"]=df['Job Title'].apply(check_title)
```
